```python
from sklearn.datasets import load_iris
iris_data = load_iris()
data = iris_data.data
targets = iris_data.target
features = ["_".join(feature.split()[0: 2]) for feature in iris_data.feature_names]
```

```python
iris_data
```

```python
import pandas as pd

iris = pd.DataFrame(iris_data.data, columns = features)
```

```python
iris
```

```python
iris['species'] = pd.Series(targets).apply(lambda val: iris_data.target_names[val])
```
