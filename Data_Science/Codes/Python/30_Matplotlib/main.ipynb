```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 200)
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.xlabel("Values of X ---->")
plt.ylabel("Sin(x) ---->")
plt.title("Graph of Sin(x) function")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
```

```python
colors = np.random.rand(50)
plt.scatter(x, y, c = colors) # We can use constance like 'b' or html color code like '#XXXXX
```

```python
sizes = 1000 * np.random.rand(50)
plt.scatter(x, y, c = colors, s = sizes)
```

```python
plt.scatter(x, y, c = colors, s = sizes, alpha = 0.5)
```

```python
plt.scatter(x, y, c = colors, s = sizes, marker = 'X')
```

```python
plt.scatter(x, y, c = colors, s = sizes, marker = 'x')
```

```python
x = np.linspace(0, 10, 200)
y = np.sin(x)
```

```python
plt.plot(x, y, '--r')
```

```python
plt.figure(figsize = (10, 4))
plt.plot(x, y, '--b')
plt.show()
```

```python
plt.plot(x, y, '--b')
plt.figure(figsize = (10, 4))
plt.xlabel("x ==>")
plt.ylabel("y ==>")
plt.show()
```

```python
plt.figure(figsize = (10, 4))
plt.plot(x, y, '--b')
plt.xlabel("x ==>")
plt.ylabel("y ==>")
plt.grid()
plt.show()
```
