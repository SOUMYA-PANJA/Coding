# SMOTE

## SMOTE Technique:
### Used for upsampling but better to use interpolation
### Connects two points in the minority class and connects a line between them and consider each point on the line as a new sample

```python
from sklearn.datasets import make_classification
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X, Y = make_classification(
                            n_samples = 1000,
                            n_features = 2,
                            n_redundant = 0,
                            n_clusters_per_class = 1,
                            weights = [0.90],
                            random_state = 10
                        )
```

```python
X.shape
```

```python
Y.shape
```

```python
df1 = pd.DataFrame(X, columns = ['f1', 'f2'])
df2 = pd.DataFrame(Y, columns = ['target'])
df = pd.concat([df1, df2], axis = 1)
```

```python
df.head()
```

```python
df['target'].value_counts()
```

```python
plt.scatter(df['f1'], df['f2'], c = df['target'])
```

### Using SMOTE

### To install in enviroment use "!pip install imblearn"

```python
from imblearn.over_sampling import SMOTE
```

```python
oversampled = SMOTE()
```

```python
X, Y = oversampled.fit_resample(df[['f1', 'f2']], df['target'])
```

```python
X.shape
```

```python
Y.shape
```

```python
df1 = pd.DataFrame(X, columns = ['f1', 'f2'])
df2 = pd.DataFrame(Y, columns = ['target'])
df = pd.concat([df1, df2], axis = 1)
```

```python
df.head()
```

```python
df['target'].value_counts()
```

```python
plt.scatter(df['f1'], df['f2'], c = df['target'])
```
