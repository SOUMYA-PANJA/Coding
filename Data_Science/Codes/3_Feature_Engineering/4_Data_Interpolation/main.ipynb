# Data Interpolation

### Data Interpolation tries to generate datapoints while fitting any curve to capture all the known datapoints

## 1. Linear Interpolation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, Y = np.arange(1, 6, dtype = 'int64'), np.arange(2, 12, 2, dtype = 'int64')
```

```python
plt.scatter(X, Y)
```

```python
x_new = np.linspace(np.min(X), np.max(X), 10)
y_new = np.interp(x_new, X, Y)
```

```python
plt.scatter(x_new, y_new)
plt.scatter(X, Y)
```

## 2. Cubic interpolation with Scipy

```python
X = np.arange(1, 6, dtype = 'int64')
Y = X**3
```

```python
from scipy.interpolate import interp1d
```

#### Create a cubic interpolation function object

```python
f = interp1d(X, Y, kind = 'cubic')
```

#### Interpolating the data

```python
x_new = np.linspace(np.min(X), np.max(X), 10)
y_new = f(x_new)
```

```python
plt.scatter(x_new, y_new)
plt.scatter(X, Y)
```

## 3. Polynomial Interpolation

```python
X = np.arange(1, 6, dtype = 'int64')
Y = X**2
```

#### Interpolate the data using polynomial interpolation
#### Fitting a 2nd degree polynomial to the data

```python
poly = np.polyfit(X, Y, 2)
```

```python
x_new = np.linspace(np.min(X), np.max(X), 10)
y_new = np.polyval(poly, x_new)
```

```python
plt.scatter(x_new, y_new)
plt.scatter(X, Y)
```
