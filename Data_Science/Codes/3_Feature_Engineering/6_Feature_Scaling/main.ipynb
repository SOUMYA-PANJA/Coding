## 1. Standardization:
### $$ z = \frac{(x - \mu)}{\sigma} $$

```python
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
total_bill = list(df['total_bill'])
```

```python
sns.histplot(total_bill)
```

```python
print(total_bill)
```

```python
mu = np.mean(total_bill)
sigma = np.std(total_bill)
```

```python
normalized_total_bill = [(i - mu) / sigma for i in total_bill]
```

```python
print(normalized_total_bill)
```

```python
sns.histplot(normalized_total_bill)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df.head()
```

```python
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(df[['total_bill', 'tip']])
```

```python
print(X)
```

## 2. Normalization:
### $$ x_{scaled} = \frac{(x - x_{min})}{x_{max} - x_{min}} $$

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X = scaler.fit_transform(df[['total_bill', 'tip']])
```

```python
print(X)
```

```python
scaler.transform([[1, 2]])
```

## 3. Unit Vector Scaling
### $$ \lVert \hat{x} \rVert = \frac{ \vec{x}}{ \lvert \vec{x} \rvert} $$

```python
from sklearn.preprocessing import normalize
```

```python
df.head()
```

```python
X = normalize(df[['total_bill', 'tip']])
```

```python
print(X)
```
