```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from tkinter import *
from tkinter import filedialog
import os
import cv2
import cmath
from matplotlib import cm
```

```python
def multi_gaussian(pos, mu, si):
  n = mu.shape[0]
  Sig_det = np.linalg.det(si)
  Sig_inv = np.linalg.inv(si)
  N = np.sqrt((2*np.pi)**n * Sig_det)
  fac = np.einsum('...k,kl,...l->...', pos-mu, Sig_inv,pos-mu)
  return np.exp(-fac/2)/N
```

```python
def gauss(x,y):
    mu = np.array([x,y])
    sd=np.std([640,480])
    Sigma = np.array([[sd,0],[0,sd]])
    X=np.linspace(0,640,640)
    Y=np.linspace(0,480,480)
    X, Y = np.meshgrid(X, Y)
    # print(X.shape+(2,))
    pos = np.empty(X.shape+(2,))
    pos[:,:,0] = X
    pos[:,:,1] = Y
    Z = multi_gaussian(pos, mu, Sigma)
    # plt.imshow(Z)
    # plt.show()
    Gaussian_Matrix.append(Z)
```

```python
global Gaussian_Matrix, All_Gaussian_Matrix, Fourier_Matrix
Gaussian_Matrix = []
All_Gaussian_Matrix = []
Fourier_Matrix = []
l = os.listdir('../coordinate/')
stat = os.listdir('../static/')
for i in range(10):
    Gaussian_Matrix = []
    f = open('../coordinate/'+l[i], 'r')
    co = f.read().split('\n')[:-1]
    for j in co:
        x, y = j.split(',')
        gauss(int(x), int(y))
    f.close()
    path = '../static/'+stat[i]
    img = cv2.imread(path, 0)
    # fln = filedialog.askopenfilename(initialdir=os.getcwd(),title="Select Image File",filetypes=(("ALL FILES","*"),("JPG file","*.jpg"),("PNG file","*.png")))
    # img = cv2.imread(fln,0)
    # cv2.imshow('image', img)
    # cv2.setMouseCallback('image',Capture_Events)
    # cv2.waitKey(0)
    # cv2.destroyAllWindows()
    M1M2arr = np.sum(Gaussian_Matrix, axis=0)
    All_Gaussian_Matrix.append(np.fft.fft2(M1M2arr))
    Fourier_Matrix.append(np.fft.fft2(img))
```
