```python
x = torch.ones(2, 2)
y = torch.ones(2, 1)
w = torch.randn(2, 1, requires_grad=True)
b = torch.randn(1, requires_grad=True)
```

```python
p = torch.sigmoid(torch.mm(x, w) + b) # prediction
loss = -y * torch.log(p) - (1 - y) * torch.log(1 - p) # cross-entropy loss
cost = loss.mean() # the cost to minimize
```

```python
cost.backward()
```

```python
print(w.grad)
print(b.grad)
```
